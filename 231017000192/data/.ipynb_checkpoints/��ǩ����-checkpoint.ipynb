```python
import jieba
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('流浪地球2影评.xlsx')
data.head()
```

```python
from snownlp import SnowNLP
# 使用Snownlp打分
def get_score(text):
    s = SnowNLP(str(text))
    return s.sentiments

data['defen'] = data['影评'].apply(get_score)
# 标签处理
def get_label(score):
    if score >= 0.5:
        return 'positive'
    else:
        return 'negative'

data['sentiment_key'] = data['defen'].apply(get_label)
```

```python
data.head()
```

```python
# 标签处理
def get_label(score):
    if score >= 0.5:
        return  1
    else:
        return  0

data['label'] = data['defen'].apply(get_label)
```
