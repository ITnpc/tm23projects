```python
import numpy as np
import pandas
import pylab as plt  # matplotlib的一个子包
from PyEMD import EMD, Visualisation  # Visualisation 可能缺引numpy
from scipy import stats
```

```python
# load the dataset
FILENAME = "data/apple_interpolate.csv"
dataframe = pandas.read_csv(FILENAME, engine='python')
dataframe = dataframe['price']
dataset = dataframe.values   # Return a Numpy representation of the DataFrame
dataset = dataset.astype('float32')  # 变为 float type
dataset = np.reshape(dataset, (dataset.shape[0]),)  
print("dataset's shape: ", dataset.shape)
print("dataset's lenth: ", len(dataset))
print(dataset)
```

```python
# Define signal，注意输入的数据结构
# t = np.linspace(0, 1, 200)
# s = np.cos(11*2*np.pi*t*t) + 6*t*t
t = np.arange(len(dataset))  # or dataset.shape[0]
s = dataset
print("s's shape: ", s.shape)
print("t's lenth: ", t.shape)
print(t)
```

```python
# Extract imfs and residue， In case of EMD
emd = EMD()
emd.emd(s)
imfs, res = emd.get_imfs_and_residue()
print(imfs.shape)
```

```python
# Plot original series
plt.plot(t, s, 'r')
plt.title("Input signal:")
plt.xlabel("Time [s]")

vis = Visualisation()
vis.plot_imfs(imfs=imfs, residue=res, t=t, include_residue=True)
# vis.plot_instant_freq(t, imfs=imfs)
vis.show()
```

```python
# 高频-低频重构（Fine-to-coarse Reconstruction）
def fine_to_coarse(imfs):
    imf_sum = np.zeros(imfs.shape[1])
    imf_fine = np.zeros(imfs.shape[1])
    imf_coarse = np.zeros(imfs.shape[1])
    df = pandas.DataFrame(index=['t', 'p'], columns=np.arange(1, len(imfs)+1))
    for n, imf in enumerate(imfs):
        imf_sum += imf
        tt, pval = stats.ttest_1samp(imf_sum, 0.0)
        print('t-statistic = %6.3f pvalue = %6.4f' % (tt, pval))
        df.iloc[0, n] = tt
        df.iloc[1, n] = pval
        if(pval < 0.05):
            index = n  # 低频从该处起
            print(n)
            imf_fine = imf_sum - imf
            imf_coarse += imf
            break
    print(imf_sum)
    for n in range(index+1, len(imfs)):
        print(n)
        imf_sum += imfs[n]
        print(imf_sum)
        tt, pval = stats.ttest_1samp(imf_sum, 0.0)
        print('t-statistic = %6.3f pvalue = %6.4f' % (tt, pval))
        df.iloc[0, n] = tt
        df.iloc[1, n] = pval
        imf_coarse += imfs[n]
    return imf_fine, imf_coarse, df
```

```python
imf_fine, imf_coarse, df = fine_to_coarse(imfs)
```
