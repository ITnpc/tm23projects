# 填补缺失的日期 & Values

```python
import pandas as pd
import utils.data_clean as tools
```

```python
# 加载数据
FILENAME = "C:/Users/demo/workplace/Time-Series-Prediction-with-LSTM/data/apple.csv"
dataframe = pd.read_csv(FILENAME)  # 会自动给行添加索引
dataframe['date'] = pd.to_datetime(dataframe['date'])  # 转换为标准日期格式
print("# shape: ", dataframe.shape)  
print(dataframe.head(30))  # 查看dataframe的前几行，默认为5
```

```python
df_filled = tools.fill_miss_dates_by_asfreq(dataframe)
print("\n 【缺失日期填充（总表部分）】df_filled's shape:", df_filled.shape, "\n", df_filled.head(30))
```

```python
# 缺失值填充
df_filled['price'] = df_filled['price'].interpolate('linear')  # 'linear': ignore the index and treat the values as equally spaced
print("\n 【价格缺失值填充（总表部分】df_insert's shape:", df_filled.shape, "\n", df_filled.head(60))
```

```python
# 结果保留两位小数
format = lambda x: "%.2f" % x
df_filled['price'] = df_filled['price'].map(format)
dir = FILENAME.split('.', 1)[0] + "_interpolate.csv"
df_filled.to_csv(dir, index=False)
print("dir: ", dir)
print("fill_miss_values done!")
```
